# 문제정의
x의 거듭제곱인 x^n의 계산

# 알고리즘 설명
result를 1로 지정한 뒤, for 반복문을 통해 n까지 x를 n번 곱한다.

# 코드개요
입력변수 : 정수 2를 500번 거듭제곱한 결과를 매개변수로 넘겨준다.

출력 : 거듭제곱한 결과값을 반환하며, 4.8 코드와 처리 시간을 계산해 출력한다.

함수설명  : 거듭제곱을 구하기 위한 함수로, 반복문을 이용해 단순 계산한다.

```python
# 알고리즘 코드
def slow_power(x,n) :    
    result = 1.0
    for i in range(n):
        result = result * x 
    return result
```

```python
# 테스트 코드
def slow_power(x,n) :    
    result = 1.0
    for i in range(n):
        result = result * x 
    return result

def power(x, n):
    if n == 0:
        return 1
    elif (n % 2) == 0:
        return power(x*x, n//2)
    else:
        return x * power(x*x, (n-1)//2)
    
import time
print("억 지 기 법 (2**500)= ", power(2.0, 500))
print("축소정복기법(2**500)= ", slow_power(2.0, 500))

t1 = time.time()
for i in range(100000) : power(2.0, 500)
t2 = time.time()
for i in range(100000) : slow_power(2.0, 500)
t3 = time.time()

print("억지기법 시간... ", t3-t2)
print("축소정복기법 시간... ", t2-t1)
```

# 복잡도 분석
n을 2의 거듭제곱 중 한 수인 2^k라고 가정하였을 경우 2^k=n 이 된다.

이때 양변에 밑이 2인 로그를 취하면 log 2 n = k가 된다. 

따라서 연산은 log 2 n에 비례하게 될것이며 시간복잡도는 O(log 2 n)
