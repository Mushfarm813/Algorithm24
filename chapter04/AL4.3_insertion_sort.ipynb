# 문제 정의
자료 배열의 모든 요소를 앞에서부터 차례대로 이미 정렬된 배열 부분과 비교하여, 자신의 위치를 찾아 삽입함으로써 정렬을 완성하는 알고리즘

# 알고리즘 설명
두 번째 자료부터 시작하여 그 앞의 값들과 비교하여 삽입할 위치를 지정한 후 값을 뒤로 옮기고 지정한 자리에 자료를 삽입하여 정렬하는 알고리즘


# 코드 개요
입력변수 : len(A), key

출력 : len(A)

함수 설명 : 입력의 크기를 지정한 후, 외부 루프는 A[i]를 key에 저장해 A[i-1]부터 검사를 하며 내부 루프는 항목들을 오른쪽으로 한 칸씩 이동한 후 위치를 왼쪽으로 이동시켜 항목 A[i]를 제 위치에 삽입시킨다

```python
# 알고리즘 코드
def insertion_sort(A):
    n=len(A)
    for i in range(1,n):
        key=A[i]
        j=i-1
        while j>=0 and A[j]>key:
            A[j+1]=A[j]
            j=j-1
        A[j+1]=key
    
```

```python
# 테스트 코드
data=[5,2,6,1,9,7,4]
print("원래의 배열 : ",data)
insertion_sort(data)
print("삽입 정렬 후의 배열 : ",data)
```

# 복잡도 분석
최선의 경우 : 리스트가 정렬되어 있을 시 내부 루프는 한번에 빠져나오고, 외부 루프는 n-1번 실행하므로 T(n)=(n-1)*1=n-1 이므로 O(n)이다.

최악의 경우 : 리스트가 반대로 정렬되어 있을 시 T=n(n-1)/2 이므로 O(n^2)이다.
