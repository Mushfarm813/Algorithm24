# 알고리즘 코드
def closest_pair(p): n = len(p) mindist = float("inf") closest = None for i in range(n-1): for j in range(i+1, n): dist = distance(p[i], p[j]) if dist < mindist: mindist = dist closest = (p[i], p[j]) return mindist

```python
# 6. 테스트 코드


import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    closest = None
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
                closest = (p[i], p[j])
    return mindist, closest

points = [(2, 5), (2, 30), (46, 30), (7,1), (12, 10), (3, 4)]

min_distance, closest_pair_coords = closest_pair(points)
print("최근접 거리:", min_distance)
print("최근접 좌표:", closest_pair_coords)
```

최근접 거리: 1.4142135623730951

최근접 좌표: ((2, 5), (3, 4))

내부와 외부의 루프가 참으로 동작하기 때문에 O(n2)이다
