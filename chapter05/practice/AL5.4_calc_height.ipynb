# 문제 정의
이진트리의 높이

# 알고리즘 설명
이진트리가 공백이면 0을 반환하고 최고 수를 구한다

# 코드개요
1. 입력 변수 : calc_height(root) :
2. 출력         : 호출문을 호출하게 되면 트리의 가장 높은 곳을 찾아서 높이가 몇인지 알려준다.
3. 함수 설명 : 왼쪽 트리의 가장 높은 곳까지 쭉 가서 더이상 자식 노드가 없다면 0을 반환해 모든곳을 둘러보고 첫 왼쪽과 오른쪽의 높이를 비교해서 제일 높은 곳의 수를 반환해준다

```python
# 알고리즘 코드
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1
```

```python
# 테스트 코드
z = TNode('Z', None, None)
i = TNode('I', None, z)
h = TNode('H', None, None)
p = TNode('P', None, None)
g = TNode('G', None, p)
d = TNode('D', g, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', h, i)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(" 트리의 높이 = ", calc_height(root))
```
