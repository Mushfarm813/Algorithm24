# 알고리즘 설명
가운데를 기준으로 왼쪽과 오른쪽을 나누고 서로 숫자를 비교하여 순차적으로 정렬한다

# 코드개요
1. 입력 변수 : quick_sort

2. 출력         : 데이터들의 숫자들을 정렬한다.

3. 함수 설명 : 가운데를 mid로 설정한 후 양쪽 끝을 피벗으로 설정한 후 그 다음 숫자랑 비교하여 순차적으로 정렬을 한다

```python
# 알고리즘 코드
# 4.12코드
def partition(A, left, right) :
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) : # 왼쪽의 작은수가 오른쪽의 큰수보다 작으면 계속 실행
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :  # 왼쪽과 오른쪽이 만나서 교차하게되면 서로 교환을 해준다
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]  # 처음에 고정된 피벗을 high 위치와 교환한다
    return high                          # 피벗의 위치로 반환

#5.3코드
def quick_sort(A, left, right) :
    if left<right :
        mid = partition(A, left, right)
        quick_sort(A, left, mid-1)
        quick_sort(A, mid+1, right)
```

```python
# 테스트 코드
data = [5, 1, 6, 2, 9, 7, 10, 13, 15, 20, 48, 3, 11, 21]
print("Original : ", data)
quick_sort(data, 0, len(data)-1)
print("QuickSort : ", data)
```

# 복잡도 분석
마스터 정리를 사용하거나 연속 대치법을 동일하게 적용하면 점근적 복잡도는 O(n log2 n)이지만

최악의 경우 트리의 높이가 n이므로, 결국 n^2에 비례하는 비교 연산이 필요하기에 최악의 경우 O(n^2)의 시간 복잡도를 가진다
