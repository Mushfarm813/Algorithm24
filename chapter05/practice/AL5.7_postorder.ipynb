# 문제 정의
이진트리의 후위 순회

# 알고리즘 설명
왼쪽부터 오른쪽으로 가는데 중위 순회와 달리 최대한 아래왼쪽에서부터 올라가며 오른쪽 노드 기준 첫 맨 아래 왼쪽부터 계산된다

# 코드 개요
1. 입력 변수 : postorder(n)

2. 출력         : 호출문으로 호출을 할 시 가장 왼쪽 서브 트리부터 처리를 하여 순차적으로 오른쪽 서브 트리까지 처리한다.

3. 함수 설명 : 중위 수식과 다른점은 중위 수식은 가장 왼쪽부터 Y축으로 순서대로 읽는다면 후위 수식은 왼쪽의 자식노드부터 읽어 올라가며 오른쪽도 똑같이 반복한다

```python
# 알고리즘 코드
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def postorder(n):
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')
```

```python
# 테스트 코드
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order : ', end=' ')
postorder(root)
print()
```
