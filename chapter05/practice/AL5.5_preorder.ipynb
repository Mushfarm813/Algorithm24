# 문제정의
이진트라의 전위 순회

# 알고리즘 설명
위에서부터 왼쪽으로 순차적으로 입력한다.

# 코드 개요
1. 입력 변수 : preorder(n)

2. 출력         : 트리구조에서 순차를 구해준다

3. 함수 설명 : 호출시 트리 구조에서 위에서 왼쪽으로 쭉 간 후에 자식노드가 없으면 한노드 위로 올라갔다 오른쪽 그다음은 또 위 노드로 갔다가 오른쪽으로 반복한다.

```python
# 알고리즘 코드
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

#5.5
def preorder(n):
    if n is not None :
        print(n.data, end=' ')
        preorder(n.left)
        preorder(n.right)
```

```python
# 테스트 코드
z = TNode('Z', None, None)
i = TNode('I', None, z)
h = TNode('H', None, None)
p = TNode('P', None, None)
g = TNode('G', None, p)
d = TNode('D', g, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', h, i)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' pre-Order : ', end='')
preorder(root)
print()
```
