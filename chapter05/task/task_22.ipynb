# 문제 : 22번

closest_pair_dist() 알고리즘을 n log n으로 개선하라. 이를 위해, strip_closest()에서 정렬 문장을 제거해야 하고, closest_pair_dist()에서 병합 정렬의 병합 기법(알고리즘 5.2)을 사용해야 할 것이다.

```python
# 알고리즘 5.9 기존 코드
import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    closest = None
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
                closest = (p[i], p[j])
    return mindist, closest

def strip_closest(P, d):
    n = len(P)
    d_min = d
    P.sort(key=lambda point: point[1])  

    for i in range(n):
        j = i + 1
        while j < n and (P[j][1] - P[i][1]) < d_min:
            dij = distance(P[i], P[j])
            if dij < d_min:
                d_min = dij
            j += 1
    return d_min

def closest_pair_dist(P):
    n = len(P)
    if n <= 3:
        return closest_pair(P)[0]  

    mid = n // 2
    mid_x = P[mid][0]

    dl = closest_pair_dist(P[:mid])
    dr = closest_pair_dist(P[mid:])
    d = min(dl, dr)

    Pm = []
    for i in range(n):
        if abs(P[i][0] - mid_x) < d:
            Pm.append(P[i])

    ds = strip_closest(Pm, d)
    return min(d, ds)

points = [(2, 5), (2, 30), (46, 30), (7, 1), (12, 10), (3, 4)]

points.sort(key=lambda point: point[0])

min_distance = closest_pair_dist(points)
print("최근접 거리:", min_distance)
```

```python
# 22번 문제를 해결한 코드
import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    closest = None
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
                closest = (p[i], p[j])
    return mindist, closest

def strip_closest(strip, d):
    n = len(strip)
    d_min = d
    strip.sort(key=lambda point: point[1]) 

    for i in range(n):
        j = i + 1
        while j < n and (strip[j][1] - strip[i][1]) < d_min:
            dij = distance(strip[i], strip[j])
            if dij < d_min:
                d_min = dij
            j += 1
    return d_min

def closest_pair_dist(P):
    n = len(P)
    if n <= 3:
        return closest_pair(P)[0]  

    mid = n // 2
    mid_x = P[mid][0]

    dl = closest_pair_dist(P[:mid]) #왼쪽 부분에 대해 분할정복
    dr = closest_pair_dist(P[mid:]) #오른쪽 부분에 대해 분할정복

    d = min(dl, dr) #양쪽 부분에서의 최소 거리를 구한다

    strip = []
    for point in P:
        if abs(point[0] - mid_x) < d:
            strip.append(point)

    ds = strip_closest(strip, d) # 최근접 점 쌍을 찾음
    return min(d, ds)

points = [(2, 5), (2, 30), (46, 30), (7, 1), (12, 10), (3, 4)]

points.sort(key=lambda point: point[0])

min_distance = closest_pair_dist(points)
print("최근접 거리:", min_distance)
```
